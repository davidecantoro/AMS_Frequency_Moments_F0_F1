# Elementi distinti ed RMSE

```python
import pandas as pd
import numpy as np

# Dataset con le Stime
poisson_data = pd.read_csv('/content/poisson_algorithm_comparison_results.csv')
uniform_data = pd.read_csv('/content/uniform_algorithm_comparison_results.csv')
exp_data = pd.read_csv('/content/exp_algorithm_comparison_results.csv')

# Valori elementi distinti
real_values_poisson = np.array([53, 61, 63, 67, 69])
real_values_uniform = np.array([199, 200, 201, 200, 200])
real_values_exp = np.array([790, 2199, 2853, 3742, 4483])

# Estrazione delle stime
ams_f0_poisson = poisson_data['AMS F0 Estimate'].values
ams_f0_n_poisson = poisson_data['AMS F0 N Estimate'].values
naive_f0_poisson = poisson_data['Naive F0 Estimate'].values

ams_f0_uniform = uniform_data['AMS F0 Estimate'].values
ams_f0_n_uniform = uniform_data['AMS F0 N Estimate'].values
naive_f0_uniform = uniform_data['Naive F0 Estimate'].values

ams_f0_exp = exp_data['AMS F0 Estimate'].values
ams_f0_n_exp = exp_data['AMS F0 N Estimate'].values
naive_f0_exp = exp_data['Naive F0 Estimate'].values

# Calcolo RMSE
rmse_ams_f0_poisson = np.sqrt(np.mean((ams_f0_poisson - real_values_poisson) ** 2))
rmse_ams_f0_n_poisson = np.sqrt(np.mean((ams_f0_n_poisson - real_values_poisson) ** 2))
rmse_naive_f0_poisson = np.sqrt(np.mean((naive_f0_poisson - real_values_poisson) ** 2))

rmse_ams_f0_uniform = np.sqrt(np.mean((ams_f0_uniform - real_values_uniform) ** 2))
rmse_ams_f0_n_uniform = np.sqrt(np.mean((ams_f0_n_uniform - real_values_uniform) ** 2))
rmse_naive_f0_uniform = np.sqrt(np.mean((naive_f0_uniform - real_values_uniform) ** 2))

rmse_ams_f0_exp = np.sqrt(np.mean((ams_f0_exp - real_values_exp) ** 2))
rmse_ams_f0_n_exp = np.sqrt(np.mean((ams_f0_n_exp - real_values_exp) ** 2))
rmse_naive_f0_exp = np.sqrt(np.mean((naive_f0_exp - real_values_exp) ** 2))

print("RMSE AMS F0 Poisson:", rmse_ams_f0_poisson)
print("RMSE AMS F0 N Poisson:", rmse_ams_f0_n_poisson)
print("RMSE Naive F0 Poisson:", rmse_naive_f0_poisson)

print("\nRMSE AMS F0 Uniforme:", rmse_ams_f0_uniform)
print("RMSE AMS F0 N Uniforme:", rmse_ams_f0_n_uniform)
print("RMSE Naive F0 Uniforme:", rmse_naive_f0_uniform)

print("\nRMSE AMS F0 Esponenziale:", rmse_ams_f0_exp)
print("RMSE AMS F0 N Esponenziale:", rmse_ams_f0_n_exp)
print("RMSE Naive F0 Esponenziale:", rmse_naive_f0_exp)
```

```python
import matplotlib.pyplot as plt

# Estrazione delle stime
ams_f0_poisson = poisson_data['AMS F0 Estimate'].values
ams_f0_n_poisson = poisson_data['AMS F0 N Estimate'].values
naive_f0_poisson = poisson_data['Naive F0 Estimate'].values

ams_f0_uniform = uniform_data['AMS F0 Estimate'].values
ams_f0_n_uniform = uniform_data['AMS F0 N Estimate'].values
naive_f0_uniform = uniform_data['Naive F0 Estimate'].values

ams_f0_exp = exp_data['AMS F0 Estimate'].values
ams_f0_n_exp = exp_data['AMS F0 N Estimate'].values
naive_f0_exp = exp_data['Naive F0 Estimate'].values

# Calcolo del RMSE
datasets = ['1000', '5000', '10000', '25000', '50000']

rmse_poisson = {
    'AMS F0': np.sqrt((ams_f0_poisson - real_values_poisson) ** 2),
    'AMS F0 N': np.sqrt((ams_f0_n_poisson - real_values_poisson) ** 2),
    'Naive F0': np.sqrt((naive_f0_poisson - real_values_poisson) ** 2)
}

rmse_uniform = {
    'AMS F0': np.sqrt((ams_f0_uniform - real_values_uniform) ** 2),
    'AMS F0 N': np.sqrt((ams_f0_n_uniform - real_values_uniform) ** 2),
    'Naive F0': np.sqrt((naive_f0_uniform - real_values_uniform) ** 2)
}

rmse_exp = {
    'AMS F0': np.sqrt((ams_f0_exp - real_values_exp) ** 2),
    'AMS F0 N': np.sqrt((ams_f0_n_exp - real_values_exp) ** 2),
    'Naive F0': np.sqrt((naive_f0_exp - real_values_exp) ** 2)
}

# Plot l'RMSE
def plot_rmse(rmse_values, title):
    plt.figure(figsize=(10, 6))
    for algorithm, rmse in rmse_values.items():
        plt.plot(datasets, rmse, label=algorithm)
    plt.xlabel('Dataset Size')
    plt.ylabel('Root Mean Squared Error (RMSE)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.savefig(title, format='svg')
    plt.show()

plot_rmse(rmse_poisson, 'RMSE_Poisson')
plot_rmse(rmse_uniform, 'RMSE_Uniforme')
plot_rmse(rmse_exp, 'RMSE_Esponenziale')
```

```python
# Stampa RMSE per ogni dataset
for dataset, i in zip(datasets, range(len(datasets))):
    print(f"\nErrori per il dataset {dataset}:")
    for algorithm in rmse_poisson.keys():
        print(f"{algorithm} - Poisson: {rmse_poisson[algorithm][i]}")
        print(f"{algorithm} - Uniforme: {rmse_uniform[algorithm][i]}")
        print(f"{algorithm} - Esponenziale: {rmse_exp[algorithm][i]}")
```
