```python
import pandas as pd
import matplotlib.pyplot as plt

# AMS
ams_f1_poisson_1000 = pd.read_csv('/kaggle/input/ams-dataset/ams_f1_poisson_75_1000.csv')
ams_f1_poisson_5000 = pd.read_csv('/kaggle/input/ams-dataset/ams_f1_poisson_75_5000.csv')
ams_f1_poisson_10000 = pd.read_csv('/kaggle/input/ams-dataset/ams_f1_poisson_75_10000.csv')
ams_f1_poisson_25000 = pd.read_csv('/kaggle/input/ams-dataset/ams_f1_poisson_75_25000.csv')
ams_f1_poisson_50000 = pd.read_csv('/kaggle/input/ams-dataset/ams_f1_poisson_75_50000.csv')

# Naive
naive_f1_poisson_1000 = pd.read_csv('/kaggle/input/ams-dataset/naive_f1_poisson_75_1000.csv')
naive_f1_poisson_5000 = pd.read_csv('/kaggle/input/ams-dataset/naive_f1_poisson_75_5000.csv')
naive_f1_poisson_10000 = pd.read_csv('/kaggle/input/ams-dataset/naive_f1_poisson_75_10000.csv')
naive_f1_poisson_25000 = pd.read_csv('/kaggle/input/ams-dataset/naive_f1_poisson_75_25000.csv')
naive_f1_poisson_50000 = pd.read_csv('/kaggle/input/ams-dataset/naive_f1_poisson_75_50000.csv')

datasets = ['1000', '5000', '10000', '25000', '50000']

# Max RSS
ams_f1_max_rss = [ams_f1_poisson_1000['max_RSS'].mean(), ams_f1_poisson_5000['max_RSS'].mean(), 
                  ams_f1_poisson_10000['max_RSS'].mean(), ams_f1_poisson_25000['max_RSS'].mean(), 
                  ams_f1_poisson_50000['max_RSS'].mean()]

naive_f1_max_rss = [naive_f1_poisson_1000['max_RSS'].mean(), naive_f1_poisson_5000['max_RSS'].mean(), 
                    naive_f1_poisson_10000['max_RSS'].mean(), naive_f1_poisson_25000['max_RSS'].mean(), 
                    naive_f1_poisson_50000['max_RSS'].mean()]

plt.figure(figsize=(10, 6))
plt.plot(datasets, ams_f1_max_rss, label='AMS F1')
plt.plot(datasets, naive_f1_max_rss, label='Naive F1')
plt.title('Max RSS - Poisson')
plt.xlabel('Dataset Size')
plt.ylabel('Max RSS (KB)')
plt.legend()
plt.savefig('poisson_f1_max_rss.svg', format='svg')
plt.show()

# Tempo di Esecuzione
ams_f1_execution_time = [ams_f1_poisson_1000['esecuzione'].mean(), ams_f1_poisson_5000['esecuzione'].mean(), 
                         ams_f1_poisson_10000['esecuzione'].mean(), ams_f1_poisson_25000['esecuzione'].mean(), 
                         ams_f1_poisson_50000['esecuzione'].mean()]

naive_f1_execution_time = [naive_f1_poisson_1000['esecuzione'].mean(), naive_f1_poisson_5000['esecuzione'].mean(), 
                           naive_f1_poisson_10000['esecuzione'].mean(), naive_f1_poisson_25000['esecuzione'].mean(), 
                           naive_f1_poisson_50000['esecuzione'].mean()]

plt.figure(figsize=(10, 6))
plt.plot(datasets, ams_f1_execution_time, label='AMS F1')
plt.plot(datasets, naive_f1_execution_time, label='Naive F1')
plt.title('Tempi di Esecuzione - Poisson')
plt.xlabel('Dataset Size')
plt.ylabel('Tempo di Esecuzione (s)')
plt.legend()
plt.savefig('poisson_f1_execution_time.svg', format='svg')
plt.show()

# Stima
ams_f1_estimate = [2**ams_f1_poisson_1000['stima'].mean(), 2**ams_f1_poisson_5000['stima'].mean(), 
                   2**ams_f1_poisson_10000['stima'].mean(), 2**ams_f1_poisson_25000['stima'].mean(), 
                   2**ams_f1_poisson_50000['stima'].mean()]

naive_f1_estimate = [naive_f1_poisson_1000['stima'].mean(), naive_f1_poisson_5000['stima'].mean(), 
                     naive_f1_poisson_10000['stima'].mean(), naive_f1_poisson_25000['stima'].mean(), 
                     naive_f1_poisson_50000['stima'].mean()]

plt.figure(figsize=(10, 6))
plt.plot(datasets, ams_f1_estimate, label='AMS F1')
plt.plot(datasets, naive_f1_estimate, label='Naive F1')
plt.title('Stime - Poisson')
plt.xlabel('Dataset Size')
plt.ylabel('Stima')
plt.legend()
plt.savefig('poisson_f1_estimate.svg', format='svg')
plt.show()

# Export CSV
df_results_poisson = pd.DataFrame({
    'Dataset Size': datasets,
    'AMS F1 Max RSS (KB)': ams_f1_max_rss,
    'Naive F1 Max RSS (KB)': naive_f1_max_rss,
    'AMS F1 Execution Time (s)': ams_f1_execution_time,
    'Naive F1 Execution Time (s)': naive_f1_execution_time,
    'AMS F1 Estimate': ams_f1_estimate,
    'Naive F1 Estimate': naive_f1_estimate
})

df_results_poisson.to_csv('/kaggle/working/poisson_f1_algorithm_comparison_results.csv', index=False)
```
